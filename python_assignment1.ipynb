3.       Compare and contrast mutable and immuatable objects in python with examples.

          revolves around whether the object can be changed after it is created.

           (a) Mutable Objects
               Definition: A mutable object is one whose state or value can be modified after it is created. Changes to a mutable object affect all references to that object.

               Examples of mutable objects:

                Lists
                Dictionaries
                Sets
              
              Example of a mutable object (list):

              python
              Copy code
              # Mutable object (list)
               my_list = [1, 2, 3]

              # Modifying the list
               my_list.append(4)

               print(my_list)  # Output: [1, 2, 3, 4]

              # The same list object is modified, and it is reflected in the original reference.
                Here, the list my_list was changed by appending a new value, and this change was reflected in all references to my_list.

           (b) Immutable Objects
               Definition: An immutable object is one whose state or value cannot be changed after it is created. Any operation that tries to modify an immutable object results in the creation of a new object.

               Examples of immutable objects:

                Integers
                Floats
                Strings
                Tuples
                Frozensets

               Example of an immutable object (tuple):

                # Immutable object (tuple)
                  my_tuple = (1, 2, 3)

4.        Discuss the different types of operators in python and provide examples of how they are used.

Answer=   In Python, operators are special symbols that perform operations on variables and values. There are several types of operators in
          Python, each serving a different purpose. These operators can be broadly categorized into the following types:

           (a) Arithmetic Operators [+, -, *, /, **, //, %]
              Arithmetic operators are used to perform basic mathematical operations.
               
              Example:

              a = 10
              b = 3

              print(a + b)  # Output: 13 (Addition)
              print(a - b)  # Output: 7  (Subtraction)
              print(a * b)  # Output: 30 (Multiplication)
              print(a / b)  # Output: 3.3333333333333335 (Division)
              print(a // b) # Output: 3  (Floor division)
              print(a % b)  # Output: 1  (Modulus)
              print(a ** b) # Output: 1000 (Exponentiation)


            (b) Comparison Operators  [==, !=, <, >, <=, >=]
                Comparison operators are used to compare two values. They return a Boolean value (True or False).

                Example:
                a = 10
                b = 5

                print(a == b)  # Output: False (Equality check)
                print(a != b)  # Output: True  (Inequality check)
                print(a > b)   # Output: True  (Greater than)
                print(a < b)   # Output: False (Less than)
                print(a >= b)  # Output: True  (Greater than or equal to)
                print(a <= b)  # Output: False (Less than or equal to)


            (c) Logical Operators  [and, or, not]
                Logical operators are used to combine conditional statements. They return a Boolean value.

                Example:
                a = 10
                b = 5

                print(a > 5 and b < 10)  # Output: True  (Both conditions are True)
                print(a > 5 or b > 10)   # Output: True  (At least one condition is True)
                print(not(a < b))        # Output: True  (Negates the condition)


            (d) Assignment Operators  [=, +=, -=, *=, /=, **=, //=, %= ]
                Assignment operators are used to assign values to variables. Python allows several shorthand operators for assignment.   

                Example:

                       a = 10
                       a += 5  # a = a + 5
                       print(a)  # Output: 15

                       a -= 3  # a = a - 3
                       print(a)  # Output: 12

                       a *= 2  # a = a * 2
                       print(a)  # Output: 24

                       a /= 4  # a = a / 4
                       print(a)  # Output: 6.0 


            (e) Bitwise Operators  [&, |, ^, ~, <<, >>]
                Bitwise operators perform operations on bits and operate at the binary level.  

                Example:
                    a = 5    # Binary: 101
                    b = 3    # Binary: 011

                    print(a & b)  # Output: 1  (Binary AND: 101 & 011 = 001)
                    print(a | b)  # Output: 7  (Binary OR: 101 | 011 = 111)
                    print(a ^ b)  # Output: 6  (Binary XOR: 101 ^ 011 = 110)
                    print(~a)     # Output: -6 (Bitwise NOT: ~101 = ...11111010)
                    print(a << 1) # Output: 10 (Left shift: 101 << 1 = 1010)
                    print(a >> 1) # Output: 2  (Right shift: 101 >> 1 = 10)         


            (f)Identity Operators  [is, is not]
               Identity operators are used to check whether two objects are the same in memory.

               Example:
                       a = [1, 2, 3]
                       b = a

                        print(a is b)       # Output: True  (Both refer to the same object)
                        print(a is not b)   # Output: False (Both refer to the same object)

                       c = [1, 2, 3]
                        print(a is c)       # Output: False (Different objects, even though they contain the same values)

            (g) Membership Operators  [in, not in]
                Membership operators are used to test whether a value or variable is found in a sequence (like a list, tuple, or string).   

                example:
                      lst = [1, 2, 3, 4, 5]

                      print(3 in lst)     # Output: True  (3 is in the list)
                      print(6 not in lst) # Output: True  (6 is not in the list)

            (h)  Ternary (Conditional) Operator  [if-else]
                 The ternary operator is a shorthand for if-else statements and is used to assign values based on a condition.                  

                 Example:
                        a = 10
                        b = 5
                        result = a if a > b else b
                        print(result)  # Output: 10 (because a > b)

5.        Explain the concept of type casting in python with with examples.

Answer= 
        Type casting is the process of converting one data type into another in Python. Python provides several built-in functions to perform type casting. These functions allow you to convert a value from one type (e.g., string, integer, float) to another type as per your needs.

        There are two types of type casting in Python:

        Implicit Type Casting (Automatic Type Conversion): This is done automatically by Python when you assign a value of one type to a variable of another type. It generally happens when a smaller data type is assigned to a larger data type, and Python ensures there is no loss of data.

        Explicit Type Casting (Manual Type Conversion): This is done manually by the programmer using Python's built-in functions such as int(), float(), str(), etc. It is used when you need to convert a variable from one data type to another explicitly.

        Implicit Type Casting Example

        x = 10       # Integer
        y = 3.14     # Float

        # Python automatically converts the integer 'x' to a float to perform addition
        result = x + y
        print(result)  # Output: 13.14

        Explicit Type Casting Examples

        Explicit type casting involves converting one data type to another using built-in functions.

         Converting a string to an integer (int())

         str_num = "100"
          int_num = int(str_num)  # Convert string to integer
          print(int_num)  # Output: 100




6.       How do conditionals statements work in python ? illustrate with example.

Answer=  Conditional Statements in Python
         Conditional statements in Python are used to execute specific blocks of code based on certain conditions. These conditions are typically boolean expressions that evaluate to either True or False. The most common conditional statements are:

         if statement: Executes a block of code if the condition is True.
         elif (else if) statement: Checks multiple conditions; executes a block of code if its condition is True and all preceding if and elif conditions are False.
         else statement: Executes a block of code if none of the if or elif conditions are True.

         Syntax:

                if condition:
                    # Code to execute if condition is True
                elif another_condition:
                    # Code to execute if another_condition is True
                else:
                    # Code to execute if no conditions above are True

                    Example : Basic if, elif, and else statements

                              age = 20

                              if age < 18:
                                  print("You are a minor.")
                              elif age >= 18 and age < 65:
                                  print("You are an adult.")
                              else:
                                  print("You are a senior citizen.") 

7.       Describe the different types of loops in python and their use cases with examples.

Answer=  In Python, loops are used to execute a block of code repeatedly as long as a condition is true. There are primarily two types of loops 
         in Python:

          for loop
          while loop
         Additionally, Python also provides break, continue, and else statements to control the flow of loops.

         (a) for Loop
             The for loop is used to iterate over a sequence (like a list, tuple, string, or range). It executes a block of code for each item in the sequence. The for loop is generally used when you know the number of iterations or need to loop through items in a sequence.

             Example 1: Iterating through a list

             fruits = ["apple", "banana", "cherry"]

             for fruit in fruits:
             print(fruit)

             output:
                    apple 
                    banana
                    cherry

        (b) while Loop
            The while loop repeats a block of code as long as a specified condition evaluates to True. This type of loop is used when you don't know the exact number of iterations beforehand, but you know the condition under which the loop should stop.   

            Example : Using while with a counter

                    count = 0

                    while count < 5:
                        print(count)
                    count += 1  # Increment the counter to avoid infinite loop 

                    output:
                           0
                           1
                           2
                           3
                           4

       Use case:
                The while loop is useful when you don't know the number of iterations in advance and need the loop to continue until a condition changes (e.g., waiting for user input, handling continuous data streams).
